```python
# 8. Custom Dictionary--------------
# Objective: Build a dictionary where users can add, remove, and search for word definitions.
# Skills: Dictionaries (without using the dict data structure), data management, search functions.
# Challenge: Implement a search functionality that matches similar words in case of typos
#  and a "history" of recently searched words.

class Node:
    def __init__(self,key,value):
        self.key = key
        self.value = value
        self.next = None

class Dict:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def add(self,key,value):
        new_node = Node(key, value)

        if not self.head:
            self.head = new_node
        else:
            self.tail.next = new_node

        self.tail = new_node
        self.size += 1

    def remove(self,key):

        current = self.head
        previous = None

        while current:
            if current.key == key:
                if previous:
                    previous.next = current.next
                else:
                    self.head = current.next
                if current == self.tail:
                    self.tail = previous
                self.size -+ 1
                print(f"{current} is removed successfully")
                return
            previous = current
            current = current.next
        print("Key not found")
        return
    
    def search(self, target):
        self.history = []
        current = self.head
        found = False
        
        while current:
            if current.key == target:
                print(f"Found: {current.key} -> {current.value}")
                self.history.append(target)
                found = True
                break
            current = current.next
            
            if not found:
                print(f"'{target}' not found.")
        print("Recent Searches:", " -> ".join(self.history) or "No recent searches")
        return
    


    def __iter__(self):
        current = self.head
        while current:
            yield current.key, current.value
            current = current.next

    def __len__(self):
        return self.size
    
    def __str__(self):
        
        elements = []
        current = self.head
        while current:
            elements.append(f"{current.key}: {current.value}")
            current = current.next
        return "{" + ":".join(elements) + "}"
    


def inpDict() :
    print("1. To add to the dictionary\n"
        "2. To remove from the dictionary\n"
        "3. To search in the dictionary\n"
        "4. To exit\n")
    while True:
        menu = int(input("Welcome! Please enter the number of the option you want:"))
        if menu == 1:
            return addInp()
        elif menu == 2:
            return removeInp()
        elif menu == 3:
            return searchInp()
        elif menu == 4:
            print("Exited")
            return
        else:
            print("Invalid Input. Please try again")

def addInp():
    add_key = input("Enter key: ")
    add_value = input("Enter value: ")
    Dict().add(add_key, add_value)
    print(f"{add_key}: {add_value} were added")


def removeInp():
    remove_key = input("Enter key to remove: ")
    Dict().remove(remove_key)
    print(f"{remove_key} is removed with its value")


def searchInp():
    search_key = input("Enter key to search: ")
    Dict().search(search_key)


# inpDict()
```
